# TotalSegmentator depuis le kernel `.venv`

Avant d'executer ce notebook, selectionner le kernel `.venv TotalSegmentator_MPS` dans Jupyter.

```python
from pathlib import Path
import sys

PROJECT_DIR = Path("/Users/thomashusson/Documents/Projets/TotalSegmentator_MPS")
EXPECTED_PYTHON = PROJECT_DIR / ".venv" / "bin" / "python"

print("Python:", sys.executable)
print("Project:", PROJECT_DIR)

if Path(sys.executable).resolve() != EXPECTED_PYTHON.resolve():
    raise RuntimeError(
        "Ce notebook doit etre lance avec le kernel '.venv TotalSegmentator_MPS'. "
        f"Interpreteur courant: {sys.executable}"
    )
```

Modifier `INPUT_DICOM` si besoin, puis executer la cellule suivante.

```python
import subprocess

INPUT_DICOM = PROJECT_DIR / "tdm_elena_savier" / "AAW_M12"

cmd = [
    sys.executable,
    "segment_abdominal_muscles.py",
    "--input-dicom",
    str(INPUT_DICOM),
    "--with-muscles",
    "--with-odiasp",
    "--with-tissue",
    "--with-total",
    "--height-cm",
    "162",
]

subprocess.run(cmd, cwd=PROJECT_DIR, check=True)
```
